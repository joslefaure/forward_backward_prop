```python
from bs4 import BeautifulSoup as bs
import requests
import warnings
warnings.filterwarnings(action='ignore')
import pandas as pd
```

```python
def content_get(url):
    res = requests.get(url,cookies={'over18': '1'},verify=False)
    soup = bs(res.text)
    date = soup.select('.article-meta-value')[3].text
    content = soup.find(id="main-content").text
    target_content = u'※ 發信站: 批踢踢實業坊(ptt.cc),'
    content = content.split(target_content)
    content = content[0].split(date)
    main_content = content[1].replace('--', '')

    return(main_content)

# function to get date, author, title, url as a pandas Dataframe
def ptt_board(url):
    ptt_url = "https://www.ptt.cc"
    res = requests.get(url = url,cookies={'over18': '1'},verify=False)
    soup = bs(res.text)
    title=[]
    url=[]
    author=[]
    date=[]
    content = []
    for infor in soup.select('.r-ent'):
        title.append(infor.select('.title')[0].text.strip())
        url.append(ptt_url+infor.find('a')['href'])
        author.append(infor.select('.author')[0].text)
        date.append(infor.select('.date')[0].text)
    ptt_content = pd.DataFrame({'title' : title,
                               'url' : url,
                               'author' : author,
                               'date' : date})
    return ptt_content
```

```python
ptt_content = ptt_board(url = "https://www.ptt.cc/bbs/Gossiping/index.html")
ptt_content.head()
```

```python
author = ptt_content.author.str.encode('utf-8').tolist()
date = ptt_content.date.str.encode('utf-8').tolist()

for i in range(0, len(ptt_content)): 
    content = content_get(ptt_content['url'][i])
    
    
    category = "Gossiping"
    
    print("********************  Latest post number: " + str(i+1) + " ***************************")
    print("Author: ", author[i])
    print("Date: ", date[i])
    print("Category: ", category)
    title = ptt_content['title'][i] # the title is not encoded properly, please refer to the dataframe above
    print("Title: ", title)
    print(content)
```
